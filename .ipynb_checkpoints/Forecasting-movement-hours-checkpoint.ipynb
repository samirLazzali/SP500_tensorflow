```python
from scipy.stats import mode
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV


from sklearn.utils import resample
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier


pd.set_option('display.max_columns', None)
warnings.filterwarnings("ignore")
```

```python
def rebalance(unbalanced_data):

    # Separate majority and minority classes
    data_minority = unbalanced_data[unbalanced_data.target==0]
    data_majority = unbalanced_data[unbalanced_data.target==1]

    # Upsample minority class
    n_samples = len(data_majority)
    data_minority_upsampled = resample(data_minority, replace=True, n_samples=n_samples, random_state=5)

    # Combine majority class with upsampled minority class
    data_upsampled = pd.concat([data_majority, data_minority_upsampled])

    data_upsampled.sort_index(inplace=True)

    # Display new class counts
    data_upsampled.target.value_counts()

    return data_upsampled
```

```python
def normalize(x):

    scaler = StandardScaler()
    x_norm = scaler.fit_transform(x.values)
    x_norm = pd.DataFrame(x_norm, index=x.index, columns=x.columns)

    return x_norm
```

```python
def get_indicators(data, n):

    hh = data['high'].rolling(n).max()
    ll = data['low'].rolling(n).min()
    up, dw = data['close'].diff(), -data['close'].diff()
    up[up<0], dw[dw<0] = 0, 0
    macd = data['close'].ewm(12).mean() - data['close'].ewm(26).mean()
    macd_signal = macd.ewm(9).mean()
    tp = (data['high'] + data['low'] + data['close']) / 3
    tp_ma = tp.rolling(n).mean()
    indicators = pd.DataFrame(data=0, index=data.index,
                              columns=['sma', 'ema', 'momentum',
                                       'sto_k','sto_d', 'rsi',
                                       'macd', 'lw_r', 'a/d', 'cci'])
    indicators['sma'] = data['close'].rolling(10).mean()
    indicators['ema'] = data['close'].ewm(10).mean()
    indicators['momentum'] = data['close'] - data['close'].shift(n)
    indicators['sto_k'] = (data['close'] - ll) / (hh - ll) * 100
    indicators['sto_d'] = indicators['sto_k'].rolling(n).mean()
    indicators['rsi'] = 100 - 100 / (1 + up.rolling(14).mean() / dw.rolling(14).mean())
    indicators['macd'] = macd - macd_signal
    indicators['lw_r'] = (hh - data['close']) / (hh - ll) * 100
    indicators['a/d'] = (data['high'] - data['close'].shift()) / (data['high'] - data['low'])
    indicators['cci'] = (tp - tp_ma) / (0.015 * tp.rolling(n).apply(lambda x: np.std(x)))
    direction = (data.close > data.close.shift()).astype(int)
    indicators['target'] = direction.shift(-1).fillna(0).astype(int)
    return indicators
```

```python
def scores(models, X, y):

    for model in models:
        y_pred = model.predict(X)
        acc = accuracy_score(y, y_pred)
        f1 = f1_score(y, y_pred)
        auc = roc_auc_score(y, y_pred)
        print("Accuracy Score: {0:0.2f} %".format(acc * 100))
        print("F1 Score: {0:0.4f}".format(f1))
        print("Area Under ROC Curve Score: {0:0.4f}".format(auc))
```

## Preprocessing
Nous allons ici créer les variables low, high et close

```python
data = data_stocks[['SP500', 'DATE', 'Days', 'Hours','Min']].set_index('DATE')
# test avec une moving average exp
#data['SP500'] = data['SP500'].ewm(20).mean()
print(data.shape)
# min max par jour
for day in data.Days.unique():
    for hours in data.Hours.unique():
        data.loc[(data.Days == day) & (data.Hours == hours), "low"] = data[(data.Days == day) & (data.Hours == hours)].min()['SP500']
        data.loc[(data.Days == day) & (data.Hours == hours), "high"] = data[(data.Days == day) & (data.Hours == hours)].max()['SP500']
data.head()
```

```python
# open close par heure
for day in data.Days.unique():
     for hours in data.Hours.unique():
            if(str(day) != "2017-07-03T00:00:00.000000000" and hours != "20"):
                sp500 = data[(data.Days == day) & (data.Hours == hours) & (data.Min == "59")]['SP500'][0]
            elif((str(day) == "2017-07-03T00:00:00.000000000"and hours == "17") and hours != "20"):
                sp500 = data[(data.Days == day) & (data.Hours == "17") & (data.Min == "00")]['SP500'][0]
            elif(str(day) != "2017-07-03T00:00:00.000000000" and hours == "20"):
                sp500 = data[(data.Days == day) & (data.Hours == "20") & (data.Min == "00")]['SP500'][0]
            data.loc[(data.Days == day) & (data.Hours == hours), "close"] = sp500
data.describe()
```

```python
data_trad = data.copy()
data_trad['Hours'] = data_trad['Hours'] + ":00:00" 
data_trad['Days'] = data_trad['Days'].astype(str) +" "+ data_trad['Hours']
data_trad['Days'] = pd.to_datetime(data_trad['Days'])
#data_trad = data_trad.set_index('Days')
data_trad = data_trad.groupby('Days').mean()
print(data_trad.shape)
data_trad
```

```python
plt.figure(figsize=(15,12))
data_trad['low'].plot()
data_trad['high'].plot()
data_trad['close'].plot()
data_trad.head(10) #2469.2100
```

```python
data_hours = get_indicators(data_trad,10).iloc[20:].drop(columns='a/d')
print(data_hours.shape)
data_hours.describe()
```

```python

val_size = 0.3
num_val = int(val_size*len(data_hours))
num_train = len(data_hours) - num_val

train = data_hours[:num_train].copy()
val = data_hours[num_train:].copy()

print('train  :', train.shape)
print('val  :', val.shape)

data_train = train
data_train = rebalance(data_train)
y = data_train.target
X = data_train.drop('target', axis=1)
X = normalize(X)

data_val = val
y_val = data_val.target
X_val = data_val.drop('target', axis=1)
X_val = normalize(X_val)


print("Quantité total de NA : ",data_train.isna().sum().sum())
data_hours.head()
# Proportion by Days : 
# (86, 11)
# train  : (61, 11)
# val  : (25, 11)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)
print("train : ",X_train.shape)
print("test : ",X_test.shape)
```

```python
models = [#GaussianNB(),
          SVC(random_state=5),
          RandomForestClassifier(random_state=5),
          MLPClassifier(random_state=5)
        ]

for model in models:
    model.fit(X_train, y_train)

scores(models, X_test, y_test)
# Accuracy Score: 51.64 %
# F1 Score: 0.5042
# Area Under ROC Curve Score: 0.5287
# Accuracy Score: 48.36 %
# F1 Score: 0.4324
# Area Under ROC Curve Score: 0.5041
# Accuracy Score: 50.00 %
# F1 Score: 0.3838
# Area Under ROC Curve Score: 0.5339
```
